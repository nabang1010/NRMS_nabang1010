```python

source = "/workspace/nabang1010/LBA_NLP/Recommendation_System/DATA/dev_small/behaviors.tsv"
behaviors = pd.read_table(source,
                              header=None,
                              names=["impression_id", "user_id", "time", "clicked_news", "impressions"])
    
behaviors.clicked_news.fillna(' ', inplace=True)
behaviors.impressions = behaviors.impressions.str.split()
behaviors
```

```python
behaviors
```

```python
user2int = {}
for row in behaviors.itertuples(index=False):
    if row.user_id not in user2int:
        user2int[row.user_id] = len(user2int) + 1
pd.DataFrame(user2int.items(), columns=['user', 'int'])
print(f'Please modify `num_users` in `src/config.py` into 1 + {len(user2int)}')
for row in behaviors.itertuples():
    behaviors.at[row.Index, 'user_id'] = user2int[row.user_id]
behaviors
```

```python
negative_sampling_ratio = 2
for row in tqdm(behaviors.itertuples(), desc="Balancing data"):
    positive = iter([x for x in row.impressions if x.endswith('1')])  # lấy các impressions kết thúc bằng 1 cho vào vòng lặp
    negative = [x for x in row.impressions if x.endswith('0')] # lấy các impressions kết thúc bằng 0 cho vào vòng lặp
    
    random.shuffle(negative) # xáo trộn các phần tử trong negative
    negative = iter(negative) # lấy các phần tử trong negative cho vào vòng lặp

    pairs = []
    try:                                                         #                                                                   
        while True:                                              #                      
            pair = [next(positive)]                              #                                      
            for _ in range(negative_sampling_ratio):             #                                                       
                pair.append(next(negative))                      # chỉ lấy 1 phàna tử positive và 2 phần tử negative ngẫu nhiên                                             
            pairs.append(pair)                                   #                                 
    except StopIteration:                                        #                            
        pass                                                     #               
    # print(pairs)                                                 #                   
    behaviors.at[row.Index, 'impressions'] = pairs               #                                                     
    # break                                            
```
