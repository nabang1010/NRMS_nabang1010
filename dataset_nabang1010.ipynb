```python
import torch
from torch.utils.data import Dataset
import pandas as pd
from ast import literal_eval
import yaml
```

```python
with open("config.yaml", "r") as f:
    config = yaml.load(f, Loader=yaml.FullLoader)
```

```python
news_path = "/workspace/nabang1010/LBA_NLP/Recommendation_System/src/NRMS_nabang1010/DATA_nabang1010/train/news_preprocessed.tsv"
news_parsed = pd.read_table(
            news_path,
            index_col='id',
            usecols=['id'] + config["NRMS_CONFIG"]["dataset_attributes"]['news'],
            converters={
                attribute: literal_eval
                for attribute in set(config["NRMS_CONFIG"]["dataset_attributes"]['news']) & set([
                    'title', 'abstract', 'title_entities', 'abstract_entities'
                ])
            })
```
